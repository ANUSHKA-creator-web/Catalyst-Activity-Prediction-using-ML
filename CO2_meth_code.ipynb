```python
# Colab-ready: VAE + Deep Wide-Residual NN pipeline (fixed: drop fully-empty cols, consistent names)
# Copy-paste entire cell into Google Colab and run.

# (Optional) pip installs (uncomment if needed)
# !pip install --quiet openpyxl joblib

# 0) Imports & reproducibility
import os, json, random
from datetime import datetime
import numpy as np, pandas as pd, matplotlib.pyplot as plt
random.seed(42); np.random.seed(42)

import tensorflow as tf
tf.random.set_seed(42)
from tensorflow.keras import layers, Model, Input
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras import backend as K # Import Keras backend

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import joblib

print("TF version:", tf.__version__)

# 1) Upload files (interactive)
from google.colab import files
print("Upload these two files when prompted:\n - Curated data_1234 datapoints.xlsx\n - Data for VAE.xlsx")
uploaded = files.upload()

# detect file names
curated_path = None
vae_path = None
for name in uploaded.keys():
    low = name.lower()
    if "curated" in low:
        curated_path = name
    if "vae" in low or "data for vae" in low or "data_for_vae" in low:
        vae_path = name

if curated_path is None or vae_path is None:
    raise SystemExit("Missing required files. Make sure both curated and VAE excel files are uploaded.")

print("Using:", curated_path, vae_path)

# 2) Load Excel sheets
df_cur = pd.read_excel(curated_path, sheet_name=0)
df_vae = pd.read_excel(vae_path, sheet_name=0)
print("curated:", df_cur.shape, "vae:", df_vae.shape)

# 3) Align rows (use min rows to avoid index mismatch)
nrows = min(len(df_cur), len(df_vae))
df_cur = df_cur.iloc[:nrows].reset_index(drop=True)
df_vae = df_vae.iloc[:nrows].reset_index(drop=True)
print("Aligned rows:", nrows)

# 4) Detect or set target
def detect_target(df):
    for c in df.columns:
        if isinstance(c, str):
            low = c.lower()
            if "sty" in low or ("space" in low and "yield" in low) or "yield" in low:
                return c
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    return numeric_cols[0] if numeric_cols else None

# Optionally override:
# target_col = "STY [mgMeOH h-1 gcat-1]"
target_col = None
if target_col is None:
    target_col = detect_target(df_cur)
if target_col is None:
    raise ValueError("No suitable numeric target found. Set target_col manually.")
print("Target:", target_col)

# 5) Separate X and y, convert target to numeric and drop rows with missing target
y_cur = pd.to_numeric(df_cur[target_col], errors='coerce')
X_cur = df_cur.drop(columns=[target_col])

# drop rows where target is NaN (and align both df_vae accordingly)
mask_valid = ~y_cur.isna()
if mask_valid.sum() < len(y_cur):
    print(f"Dropping {len(y_cur)-mask_valid.sum()} rows with NaN target.")
    y_cur = y_cur[mask_valid].reset_index(drop=True)
    X_cur = X_cur[mask_valid].reset_index(drop=True)
    df_vae = df_vae[mask_valid].reset_index(drop=True)
    nrows = len(y_cur)

# 6) Numeric-only features & drop fully-empty numeric columns
X_cur_num = X_cur.select_dtypes(include=[np.number]).copy()
VAE_num = df_vae.select_dtypes(include=[np.number]).copy()

X_cur_num = X_cur_num.dropna(axis=1, how='all')   # drop fully empty numeric columns
VAE_num = VAE_num.dropna(axis=1, how='all')

print("Numeric feature shapes (curated, vae):", X_cur_num.shape, VAE_num.shape)

# 7) Imputation (mean)
imputer = SimpleImputer(strategy="mean")
X_cur_imp = pd.DataFrame(imputer.fit_transform(X_cur_num))
VAE_imp = pd.DataFrame(imputer.fit_transform(VAE_num))

print("After imputation shapes:", X_cur_imp.shape, VAE_imp.shape)

# 8) VAE: build, train, extract latent features
input_dim = VAE_imp.shape[1]
latent_dim = max(4, min(16, input_dim // 2))
encoder_hidden = max(32, input_dim * 2)

scaler_vae = StandardScaler()
VAE_scaled = scaler_vae.fit_transform(VAE_imp)

# encoder
vae_inputs = Input(shape=(input_dim,), name="vae_input")
h = layers.Dense(encoder_hidden, activation="relu")(vae_inputs)
h = layers.Dense(max(encoder_hidden//2, 32), activation="relu")(h)
z_mean = layers.Dense(latent_dim, name="z_mean")(h)
z_log_var = layers.Dense(latent_dim, name="z_log_var")(h)

def sampling(args):
    z_m, z_lv = args
    eps = tf.random.normal(shape=tf.shape(z_m))
    return z_m + tf.exp(0.5 * z_lv) * eps

z = layers.Lambda(sampling)([z_mean, z_log_var])

# decoder
d = layers.Dense(max(encoder_hidden//2, 32), activation="relu")(z)
d = layers.Dense(encoder_hidden, activation="relu")(d)
vae_outputs = layers.Dense(input_dim, activation=None)(d)

# Custom VAE loss layer
class VAELossLayer(layers.Layer):
    def __init__(self, **kwargs):
        super(VAELossLayer, self).__init__(**kwargs)

    def call(self, inputs):
        x, x_decoded, z_m, z_lv = inputs
        recon_loss = K.sum(K.square(x - x_decoded), axis=-1)
        kl_loss = -0.5 * K.sum(1 + z_lv - K.square(z_m) - K.exp(z_lv), axis=-1)
        total_loss = K.mean(recon_loss + kl_loss)
        self.add_loss(total_loss)
        return x_decoded # Return the output of the VAE decoder

vae = Model(vae_inputs, VAELossLayer()([vae_inputs, vae_outputs, z_mean, z_log_var]), name="VAE")

vae.compile(optimizer=tf.keras.optimizers.Adam(1e-3))
vae.summary()

# Train VAE with early stopping
vae_ckpt = "/content/vae_best.keras"
es_vae = EarlyStopping(monitor="loss", patience=20, restore_best_weights=True)
mc_vae = ModelCheckpoint(vae_ckpt, monitor="loss", save_best_only=True)
EPOCHS_VAE = 200
history_vae = vae.fit(VAE_scaled, VAE_scaled, epochs=EPOCHS_VAE, batch_size=32,
                      callbacks=[es_vae, mc_vae], verbose=1)

# encoder model (z_mean)
encoder = Model(vae_inputs, z_mean, name="encoder_mean")
latents = encoder.predict(VAE_scaled, batch_size=128)
latent_df = pd.DataFrame(latents, columns=[f"vae_latent_{i}" for i in range(latents.shape[1])])
print("Latent features shape:", latent_df.shape)

# 9) Combine curated features + VAE latent features
X_combined = pd.concat([X_cur_imp.reset_index(drop=True), latent_df.reset_index(drop=True)], axis=1)
print("Combined feature shape:", X_combined.shape)

# Convert column names to strings before splitting and scaling
X_combined.columns = X_combined.columns.astype(str)

# 10) Train/test split and scaling
X_train, X_test, y_train, y_test = train_test_split(X_combined, y_cur, test_size=0.2, random_state=42)
scaler_feat = StandardScaler()
X_train_s = scaler_feat.fit_transform(X_train)
X_test_s = scaler_feat.transform(X_test)

# 11) Build deep Wide+Deep+Residual network
def residual_block(x_in, units):
    x = layers.Dense(units)(x_in)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.Dense(units)(x)
    if x_in.shape[-1] != units:
        shortcut = layers.Dense(units)(x_in)
    else:
        shortcut = x_in
    x = layers.Add()([x, shortcut])
    x = layers.Activation("relu")(x)
    return x

inp = Input(shape=(X_train_s.shape[1],))
x = layers.Dense(512, activation='relu')(inp)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.2)(x)
x = residual_block(x, 512)
x = layers.Dropout(0.2)(x)
x = residual_block(x, 512)
x = layers.Dropout(0.2)(x)

deep = layers.Dense(256, activation='relu')(x)
deep = layers.BatchNormalization()(deep)
deep = layers.Dense(128, activation='relu')(deep)

wide = layers.Dense(128, activation='relu')(inp)
combined = layers.Concatenate()([wide, deep])
combined = layers.Dense(128, activation='relu')(combined)
combined = layers.BatchNormalization()(combined)
combined = layers.Dropout(0.2)(combined)
out = layers.Dense(1, activation='linear')(combined)

model = Model(inp, out, name="WideDeepResNet")
model.compile(optimizer=tf.keras.optimizers.Adam(1e-3), loss='mse', metrics=['mae'])
model.summary()

# 12) Train NN with callbacks
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
model_ckpt = f"/content/deep_model_best_{timestamp}.keras"
es = EarlyStopping(monitor="val_loss", patience=25, restore_best_weights=True)
mc = ModelCheckpoint(model_ckpt, monitor="val_loss", save_best_only=True)
rlr = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=8, min_lr=1e-6)

history_nn = model.fit(X_train_s, y_train, validation_split=0.15, epochs=400, batch_size=32,
                       callbacks=[es, mc, rlr], verbose=1)

# 13) Evaluate & save outputs
# Predict
y_pred = model.predict(X_test_s).flatten()

# Evaluate metrics
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

metrics = {
    "r2": float(r2_score(y_test, y_pred)),
    "mae": float(mean_absolute_error(y_test, y_pred)),
    "rmse": float(rmse)
}

print("Test metrics:", metrics)
# Make artifacts folder
outdir = f"/content/artifacts_{timestamp}"
os.makedirs(outdir, exist_ok=True)

# Save models & scalers & predictions
vae.save(os.path.join(outdir, "vae_full.keras"))
encoder.save(os.path.join(outdir, "vae_encoder.keras"))
model.save(os.path.join(outdir, "nn_best.keras"))
joblib.dump(scaler_vae, os.path.join(outdir, "scaler_vae.pkl"))
joblib.dump(scaler_feat, os.path.join(outdir, "scaler_feat.pkl"))
joblib.dump(imputer, os.path.join(outdir, "imputer_mean.pkl"))
pd.Series(list(X_combined.columns)).to_csv(os.path.join(outdir, "feature_columns.csv"), index=False)

pred_df = pd.DataFrame({"y_true": y_test.values, "y_pred": y_pred})
pred_df.to_csv(os.path.join(outdir, "predictions_test.csv"), index=False)

with open(os.path.join(outdir, "run_report.json"), "w") as f:
    json.dump({
        "timestamp": timestamp,
        "rows_used": int(nrows),
        "target_col": target_col,
        "feature_count": int(X_combined.shape[1]),
        "vae_latent_dim": int(latents.shape[1]),
        "metrics": metrics
    }, f, indent=2)

# 14) Plots
plt.figure(figsize=(8,5))
plt.plot(history_nn.history['loss'], label='train_loss')
plt.plot(history_nn.history['val_loss'], label='val_loss')
plt.title("NN train & val loss")
plt.xlabel("epoch"); plt.ylabel("mse"); plt.legend(); plt.grid(True)
plt.savefig(os.path.join(outdir, "nn_loss.png"))
plt.show()

plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred, alpha=0.6)
mn, mx = min(y_test.min(), y_pred.min()), max(y_test.max(), y_pred.max())
plt.plot([mn,mx],[mn,mx],'k--'); plt.xlabel("True"); plt.ylabel("Pred"); plt.title("True vs Pred")
plt.savefig(os.path.join(outdir, "true_vs_pred.png"))
plt.show()

print("Artifacts saved to:", outdir)
print("Files: ", os.listdir(outdir)[:20])
```

```python
# Enhanced visualization suite for Catalytic Activity predictions
# Saves: CatalyticActivity_Enhanced_Visualizations.png and CatalyticActivity_Predictions_Enhanced.csv

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import scipy.stats as stats

# plotting defaults (your preferred sizes)
TITLE_FS = 22
AXIS_TITLE_FS = 22
AXIS_TICKS_FS = 18
LEGEND_FS = 18
sns.set(style="whitegrid")

# --- Load predictions/results robustly ---
try:
    # prefer variables already present in session
    y_test  # just to test if exists
    have_vars = True
except Exception:
    have_vars = False

if have_vars and ('y_test' in globals()) and ('y_pred' in globals()):
    y_test_arr = np.asarray(y_test).flatten()
    y_pred_arr = np.asarray(y_pred).flatten()
    # if you also produced cumulative or extra columns earlier, ignore — we recompute
else:
    # try to load CSV that we saved earlier
    csv_candidates = [
        "CatalyticActivity_Predictions_Full.csv",
        "CatalyticActivity_Predictions.csv",
        "CatalyticActivity_Predictions_Enhanced.csv"
    ]
    found = None
    for c in csv_candidates:
        if os.path.exists(c):
            found = c
            break
    if found is None:
        raise FileNotFoundError("No y_test/y_pred variables found and no predictions CSV present. "
                                "Please run predictions or place 'CatalyticActivity_Predictions_Full.csv' in working dir.")
    df = pd.read_csv(found)
    # Try to infer columns
    if "Actual" in df.columns and "Predicted" in df.columns:
        y_test_arr = df["Actual"].to_numpy()
        y_pred_arr = df["Predicted"].to_numpy()
    else:
        # fallback: first two numeric columns
        numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
        if len(numeric_cols) >= 2:
            y_test_arr = df[numeric_cols[0]].to_numpy()
            y_pred_arr = df[numeric_cols[1]].to_numpy()
        else:
            raise ValueError(f"CSV {found} doesn't contain suitable Actual/Predicted columns.")

# compute diagnostics
residuals = y_test_arr - y_pred_arr
mse = mean_squared_error(y_test_arr, y_pred_arr)
rmse = float(np.sqrt(mse))
mae = float(mean_absolute_error(y_test_arr, y_pred_arr))
r2 = float(r2_score(y_test_arr, y_pred_arr))

# identify top absolute errors (for annotation)
abs_err = np.abs(residuals)
topk = 5
top_idx = np.argsort(-abs_err)[:topk]

# prepare cumulative residuals
cumulative_error = np.cumsum(residuals)

# Try get feature importance if model present
feature_imp = None
feature_names = None
if 'model' in globals():
    try:
        fi = getattr(model, "feature_importances_", None)
        if fi is not None:
            # if you used a dataframe for training, try to get column names
            if 'X_test_s' in globals():
                try:
                    feature_names = list(X_test_s.columns)
                except Exception:
                    feature_names = [f"f{i}" for i in range(len(fi))]
            else:
                feature_names = [f"f{i}" for i in range(len(fi))]
            feature_imp = np.array(fi)
    except Exception:
        feature_imp = None

# create figure
fig = plt.figure(figsize=(20, 18))
fig.suptitle("Enhanced Catalytic Activity Prediction Diagnostics", fontsize=TITLE_FS, y=0.98)

# Ax1: Actual vs Predicted (main) with regression line and metrics
ax1 = plt.subplot2grid((4, 3), (0, 0), colspan=2, rowspan=2)
sns.scatterplot(x=y_test_arr, y=y_pred_arr, ax=ax1, s=40, alpha=0.7)
# ideal line
mn = min(np.nanmin(y_test_arr), np.nanmin(y_pred_arr))
mx = max(np.nanmax(y_test_arr), np.nanmax(y_pred_arr))
ax1.plot([mn, mx], [mn, mx], linestyle='--', color='red', linewidth=2, label='Ideal')
# regression line
sns.regplot(x=y_test_arr, y=y_pred_arr, ax=ax1, scatter=False, lowess=False, truncate=True, line_kws={'color':'black', 'linewidth':1.5})
# annotate top errors
for i in top_idx:
    ax1.annotate(f"{int(i)}", (y_test_arr[i], y_pred_arr[i]), textcoords="offset points", xytext=(6,6), fontsize=12, color='darkred')

ax1.set_xlabel("Actual Catalytic Activity", fontsize=AXIS_TITLE_FS)
ax1.set_ylabel("Predicted Catalytic Activity", fontsize=AXIS_TITLE_FS)
ax1.set_title("Actual vs Predicted (with regression, top-error indices labeled)", fontsize=18)
ax1.tick_params(axis='both', labelsize=AXIS_TICKS_FS)
ax1.legend(fontsize=LEGEND_FS)
# metrics text box
metrics_text = f"R² = {r2:.3f}\nMAE = {mae:.2f}\nRMSE = {rmse:.2f}\nN = {len(y_test_arr)}"
ax1.text(0.02, 0.98, metrics_text, transform=ax1.transAxes, fontsize=14, verticalalignment='top',
         bbox=dict(boxstyle="round,pad=0.4", fc="wheat", alpha=0.6))

# Ax2: Residual distribution (hist + KDE)
ax2 = plt.subplot2grid((4, 3), (0, 2), colspan=1, rowspan=1)
sns.histplot(residuals, bins=40, kde=True, ax=ax2, edgecolor='k')
ax2.set_title("Residuals Distribution (hist + KDE)", fontsize=16)
ax2.set_xlabel("Residual (Actual - Predicted)", fontsize=AXIS_TITLE_FS-4)
ax2.tick_params(axis='both', labelsize=AXIS_TICKS_FS-4)
# annotate skew/kurtosis
skew = stats.skew(residuals, nan_policy='omit')
kurt = stats.kurtosis(residuals, nan_policy='omit')
ax2.text(0.02, 0.95, f"Skew={skew:.2f}\nKurt={kurt:.2f}", transform=ax2.transAxes, fontsize=12,
         verticalalignment='top', bbox=dict(boxstyle="round,pad=0.3", fc="lightgrey", alpha=0.5))

# Ax3: Residuals vs Actual with LOWESS/reg smoothing
ax3 = plt.subplot2grid((4, 3), (1, 2), colspan=1, rowspan=1)
sns.scatterplot(x=y_test_arr, y=residuals, ax=ax3, s=35, alpha=0.6)
# LOWESS via seaborn regplot lowess=True
sns.regplot(x=y_test_arr, y=residuals, ax=ax3, scatter=False, lowess=True, line_kws={'color':'red', 'linewidth':1.5})
ax3.axhline(0, color='black', linestyle='--')
ax3.set_title("Residuals vs Actual (LOWESS smoothed)", fontsize=16)
ax3.set_xlabel("Actual Catalytic Activity", fontsize=AXIS_TITLE_FS-4)
ax3.set_ylabel("Residuals", fontsize=AXIS_TITLE_FS-4)
ax3.tick_params(axis='both', labelsize=AXIS_TICKS_FS-4)

# Ax4: Residuals over sample index (line) and highlight top absolute errors
ax4 = plt.subplot2grid((4, 3), (2, 0), colspan=1, rowspan=1)
ax4.plot(residuals, marker='o', linestyle='-', markersize=3, alpha=0.6)
ax4.axhline(0, color='red', linestyle='--')
ax4.set_title("Residuals over Sample Index", fontsize=16)
ax4.set_xlabel("Sample Index", fontsize=AXIS_TITLE_FS-4)
ax4.set_ylabel("Residuals", fontsize=AXIS_TITLE_FS-4)
ax4.tick_params(axis='both', labelsize=AXIS_TICKS_FS-4)
# highlight top errors
ax4.scatter(top_idx, residuals[top_idx], color='red', s=60, label='Top abs errors')
ax4.legend(fontsize=12)

# Ax5: Cumulative error
ax5 = plt.subplot2grid((4, 3), (2, 1), colspan=1, rowspan=1)
ax5.plot(cumulative_error, linestyle='-', linewidth=1.2)
ax5.set_title("Cumulative Residuals (Bias over dataset)", fontsize=16)
ax5.set_xlabel("Sample Index", fontsize=AXIS_TITLE_FS-4)
ax5.set_ylabel("Cumulative Residuals", fontsize=AXIS_TITLE_FS-4)
ax5.tick_params(axis='both', labelsize=AXIS_TICKS_FS-4)
ax5.axhline(0, color='black', linestyle='--')

# Ax6: Q-Q plot of residuals
ax6 = plt.subplot2grid((4, 3), (2, 2), colspan=1, rowspan=1)
stats.probplot(residuals, dist="norm", plot=ax6)
ax6.set_title("Q-Q Plot of Residuals", fontsize=16)
ax6.tick_params(axis='both', labelsize=AXIS_TICKS_FS-4)

# Ax7: Log-transformed Actual vs Predicted (if all positive)
ax7 = plt.subplot2grid((4, 3), (3, 0), colspan=1, rowspan=1)
log_ok = np.all(y_test_arr > 0) and np.all(y_pred_arr > -1e-9)
if log_ok:
    ly = np.log1p(y_test_arr)
    lp = np.log1p(y_pred_arr)
    sns.scatterplot(x=ly, y=lp, ax=ax7, s=35, alpha=0.6)
    ax7.plot([min(ly), max(ly)], [min(ly), max(ly)], linestyle='--', color='red')
    sns.regplot(x=ly, y=lp, ax=ax7, scatter=False, lowess=False, line_kws={'color':'black'})
    ax7.set_title("Log(1+value) Actual vs Predicted", fontsize=16)
    ax7.set_xlabel("log(1+Actual)", fontsize=AXIS_TITLE_FS-6)
    ax7.set_ylabel("log(1+Predicted)", fontsize=AXIS_TITLE_FS-6)
    ax7.tick_params(axis='both', labelsize=AXIS_TICKS_FS-6)
else:
    ax7.text(0.5, 0.5, "Log-transform not applied (non-positive values present)", horizontalalignment='center',
             verticalalignment='center', fontsize=12)
    ax7.axis('off')

# Ax8: Prediction calibration (mean predicted vs mean actual by bin)
ax8 = plt.subplot2grid((4, 3), (3, 1), colspan=1, rowspan=1)
bins = 10
df_tmp = pd.DataFrame({"actual": y_test_arr, "pred": y_pred_arr})
df_tmp['bin'] = pd.qcut(df_tmp['actual'], q=bins, duplicates='drop')
calib = df_tmp.groupby('bin').agg({'actual':'mean', 'pred':'mean'}).reset_index()
sns.scatterplot(x='actual', y='pred', data=calib, ax=ax8)
ax8.plot([calib.actual.min(), calib.actual.max()], [calib.actual.min(), calib.actual.max()], '--', color='red')
ax8.set_xlabel("Mean Actual (bin)", fontsize=AXIS_TITLE_FS-6)
ax8.set_ylabel("Mean Predicted (bin)", fontsize=AXIS_TITLE_FS-6)
ax8.set_title("Calibration by Actual Bins", fontsize=16)
ax8.tick_params(axis='both', labelsize=AXIS_TICKS_FS-6)

# Ax9: Feature importance or placeholder
ax9 = plt.subplot2grid((4, 3), (3, 2), colspan=1, rowspan=1)
if feature_imp is not None:
    idx = np.argsort(feature_imp)
    ax9.barh(np.array(feature_names)[idx], feature_imp[idx])
    ax9.set_title("Feature Importance", fontsize=16)
    ax9.tick_params(axis='both', labelsize=AXIS_TICKS_FS-6)
else:
    ax9.text(0.5, 0.5, "Feature Importance not available for this model", horizontalalignment='center',
             verticalalignment='center', fontsize=14)
    ax9.axis('off')

plt.tight_layout(rect=[0, 0.03, 1, 0.96])
out_fig = "CatalyticActivity_Enhanced_Visualizations.png"
plt.savefig(out_fig, dpi=300)
plt.show()

# Save results CSV with diagnostics
out_df = pd.DataFrame({
    "Actual": y_test_arr,
    "Predicted": y_pred_arr,
    "Residual": residuals,
    "AbsError": abs_err,
    "CumulativeResidual": cumulative_error
})
out_csv = "CatalyticActivity_Predictions_Enhanced.csv"
out_df.to_csv(out_csv, index=False)

print(f"Saved figure -> {out_fig}")
print(f"Saved predictions CSV -> {out_csv}")
print(f"Metrics: R2={r2:.4f}, MAE={mae:.4f}, RMSE={rmse:.4f}")

```
